# Extract information from multiple documents


```python
file_names = os.listdir(os.path.join("datasets","co_presentations"))

fdp = FinancialDocumentProcessor(data_dir=os.path.join("datasets","co_presentations"))

for file_name in file_names:
    try:
        print(f"Processing {file_name}...")
        fdp.ingest_extract_save(file_name)
        print(f"Successfully processed {file_name}.")

    except Exception as e:
        print(f"Error processing {file_name}: {e}")
        continue
```
